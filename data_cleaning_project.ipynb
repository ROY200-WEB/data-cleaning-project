```python
import pandas as pd
```

```python
data = {
    "ID": [101, 102, 103, 104, 104, 106, None],
    "Name": ["Srikar", "Rahul", "Ananya", "Priya", "Priya", None, "Kiran"],
    "Age": [21, 22, None, 24, 24, 23, "Twenty"],
    "Join_Date": ["2024-01-15", "15/02/2024", "2024/03/10", "April 5 2024", "April 5 2024", None, "2024-13-01"],
    "Salary": [25000, 30000, 28000, None, None, 32000, 27000]
}

df = pd.DataFrame(data)

df
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()

df
```

```python
df["Age"] = df["Age"].replace("Twenty", 20)

df["Age"] = pd.to_numeric(df["Age"])

df["Age"].fillna(df["Age"].mean(), inplace=True)

df["Salary"].fillna(df["Salary"].mean(), inplace=True)

df["Name"].fillna("Unknown", inplace=True)

df
```

```python
df["Join_Date"] = pd.to_datetime(df["Join_Date"], errors="coerce")

df
```

```python
df["Join_Date"].fillna("2024-01-01", inplace=True)

df
```

```python
df["ID"].fillna(107, inplace=True)

df
```

```python
print("Missing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())

print("\nDataset Info:")
print(df.info())
```

```python
df.to_csv("cleaned_dataset.csv", index=False)

print("Final cleaned dataset saved successfully!")
```
